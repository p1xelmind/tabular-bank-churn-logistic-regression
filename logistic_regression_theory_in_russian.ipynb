#### Создание изолированного виртуального окружения (venv)
Прежде, чем приступить к исследованию, необходимо создать изолированное виртуальное окружение (venv). Это важно сделать еще до реализации Jupyter Notebooks, чтобы блоки кода в ipynb выполнялись с использованием библиотек из виртуального окружения.    

Перейдем в терминале в корневую папку проекта и введем команду `python -m venv venv`. Данная команда состоит из трех частей. `python` запускает сам Python; `-m venv` - встроенный модуль (пакет), который отвечает за создание окружения, флаг `-m` расшифровывается как module, он приказывает интерпретатору Python найти в своих системных папках указанный модуль и запустить его как самостоятельную программу; `venv` - имя папки, куда установятся все файлы библиотек, можно назвать как угодно, к примеру, project_env вместо venv.  

После этого в корне проекта появится папка `venv`. В ней мы найдем несколько разных папок: Include, Lib, Scripts и файл pyvenv.cfg. Папка Include изначально пустая, она нужна лишь тогда, когда устанавливаются библиотеки, написанные на языке C, которым требуется компиляция при сборке (например, некоторые компоненты numpy). Папка Lib является хранилищем всех библиотек. В нее устанавливаются все новые библиотеки, которые скачиваются через `pip install`. Они "складываются" в подпапку "site-packages". Папка Scripts содержит в себе исполняемые файлы. Например, в ней находятся `python.exe`, который являетс копией глобального Python. А также `pip.exe`. Также в данной папке хранятся скрипты активации. Файл `pyvenv.cfg` является конфигурационным файлом окружения, он содержит путь к исходному (глобальному) Python, точную версию Python и флаг `include-system-site-packages = false`, который запрещает окружению видить библиотеки из глобальной среды.   

После создания папки venv, ее необходимо активировать. На Windows 11 это делается с помощью команды `venv/Scripts/activte`. Вообще, правильной командой для CMD является `venv\Scripts\activate.bat`, а для PowerShell `.\venv\Scripts\Activate.ps1`. Но сработает и команда `venv/Scripts/activate`. PowerShell достаточно гибкий, поэтому он понимает и обратные слеши и расширения нужные сам может найти.    
После активации venv, в терминале перед путем корневой папки проекта должно светиться зеленым строка `(venv)`. Это значит, что виртуальное окружение активно.

#### Составление файла .gitignore
Теперь необходимо создать файл `.gitignore`. Он необходим для того, чтобы не тянуть в GitHub мусор. Весь файл `.gitignore` будет выглядеть так:
```
# Jupyter Notebook
.ipynb_checkpoints/
*.ipynb~
*.ipynb.orig

# Python
__pycache__/
*.py[cod]
*.so
.env
venv/
env/

# IDEs
.vscode/
.idea/

# Kaggle data
data/
*.csv

# Misc
.DS_Store
```
Точка в начале означает, что после точки идет расширение. Если на конце стоит слеш `/`, то это папка (директория). И если в одной строке присутствует и точка и слеш, то значит, что это скрытая директория. Звездочка `*` перед точкой означает, что имя файла может быть любым.  

Теперь разберем каждую секцию. **Первая секция** посвящена Jupyter Notebooks. `.ipynb_checkpoints/` - это скртыая папка, которую создает Jupyter автоматически для автосохранения каждые несколько минут. Внутри этой папки JSON-файлы вида `notebook_name-checkpoint.ipynb`. Это резервные копии на случай сбоя браузера или ядра. Они полностью дублируют содержимое основного ноутбука и не несут ценности для версионирования. `*.ipynb~` является временным файлом автосохранения или резервной копией. Файлы такого формата создаются Jupyter Notebook или текстовым редактором. Они нужны, если редактор или браузер внезапно закроется, этот файл позволит восстановить несохраненные изменения. `*.ipynb.orig` - оргигинальная версия файла до какого-либо автоматического изменения. Это страховка. Перед тем как модифицировать ноутбук, сохраняется его точная копия с расширением `.orig`. Если что-то пойдет не так (сломается JSON-структура, потеряются ячейки), то можно переименовать .orig обратно в .ipynb и вернуть всё как было.  

В общем, эти все файлы - это резервные копии ноутбуков. И единственная разница между ними заключается в том, что эти копии создаются при разных условиях.   

**Вторая секция** относится к файлам Python. `__pycache__/` является папкой, в которой хранится скомпилированный байт-код (pyc-файлы), а `*.py[cod]` означает скомпилированные файлы, хранящиеся в папке `__pycache__/`. `[cod]` означает любой элемент из скобок. Так пишется, чтобы не писать сразу 3 расширение, а обойтись одним. `*.pyc` -  compiled (скомпилированный байт-код); `*.pyo` - optimized (оптимизированный байт-код, создавался при запуске с флагом -O в старых версиях Python); `*.pyd` - dynamic (динамические библиотеки на C/C++ для Windows, аналог .so на Linux). В Python 3.x файлы `.pyo` больше не создаются, однако `*.py[cod]` является исторически сложившимся шаблоном, поэтому в .gitignore продолжают писать именно так. В общем, `*.py[cod]` - это скомпилированные файлы, которые хранятся в папке `__pycache__/`. `*.so` - это, как было написано ранее, аналог `*.pyd` для Linux и Mac. `.env` - это файл с переменными окружения (API-ключи, пароли к БД). `venv/` и `env/` - это папки виртального окружения, которые мы настраивали ранее. Они содержат тысячи файлов, специфичных для конкретной ОС и архитектуры. Воспроизводимость обеспечивается через requirements.txt, а не через копирование папки venv. Папка именно 2, так как venv могут называться по разному. При желании можно даже добавить еще `.venv/`.   

**Третья секция** посвящена файлам среды разработки. `.vscode/` - это папка, содержащая файлы настройки VSCode. В ней содержатся найстройки шрифтроф, тем, расширений. Коммитить это в GitHub - плохой тон. `.idea/` аналогична vscode. Только используется PyCharm/IntelliJ IDEA.   

**Четвертая секция** отведена для файлов Kaggle. Все датасеты с Kaggle имеют расширение `*.csv`. И как правило, данные файлы помещают в папку с названием `data/`. Датасеты весят десятки и даже сотни мегабайт. GitHub имеет лимит 100 МБ на файл и не предназначен для хранения данных. Данные должны скачиваться отдельно (через Kaggle API или вручную). `*.csv` является страховкой на тот случай, если датасет лежит в корне, а не в папке.  

И заключительная, **пятая секция** отведена для Misc (miscellaneous, что переводится, как разное, прочее). В структуре .gitignore эта секция используется как "свалка" для файлов, которые не относятся к конкретным категориям (Python, IDE, Data), являются системными артефактами операционных систем, и их слишком мало, чтобы выделять под них отдельный заголовок. `.DS_Store` (Desktop Services Store) - это скрытый системный файл, который автоматически создаётся только в macOS. Файл хранит пользовательские настройки отображения папки в Finder: расположение иконок, размер иконок, режим просмотра (список, сетка, колонки), фоновое изображение папки, сортировку и порядок файлов. Каждый раз, когда открывается папка в Finder или меняется её вид, macOS создаёт или обновляет этот файл. Он появляется даже на внешних дисках и сетевых хранилищах. Хоть и на Windows такого файла не появится, все равно стоит помещать его в .gitignore.   

#### Загрузка проекта на GitHub
Для начала требуется инициализировать репозиторий Git в текущей директории. Для этого переходим в терминале в папку проекта и вводим команду `git init`, она превращает обычную папку с файлами в проект под управлением системы контроля версий Git. Данная команда в корне папки создает скрытую папку `.git/`. Внутри нее формируются служебные файлы и подпапки, необходимые для работы Git. А также создается начальная ветка (по умолчанию `master` или `main`, в зависимости от конфигурации Git). То есть на данном этапе Git знает о существовании проекта, но не отслеживает ни единого файла. При этом не обязательно находиться именно в venv. Активация venv нужна только для команд python, pip и запуска Jupyter.    

После этого вводим команду `git add .`. Она добавляет все изменения из текущей директории в индекс. Точка означает "текущая директория и всё, что внутри неё рекурсивно". Данная команда не загружает файлы в репозиторий (в базу данных Git), а лишь "помечает" их как "готовые к сохранению". И тут важен файл `.gitignore`, который был реализован ранее. Все файлы, которые присутствуют в `.gitignore` не попадают в индекс.    

Далее вводим команду `git status`. Она показывает текущее состояние рабочей директории и индекса. Команда показывает: какие файлы изменены, но не добавлены в индекс (красные); какие файлы добавлены в индекс и готовы к коммиту (зеленые); есть ли untracked файлы; на какой ветке находимся в данный момент. Данная команда нужна для того, чтобы определить, не коммитим ли мы что-то лишнее.  

После этого используется команда `git commit -m "Initial commit"`. Она берет все файлы, находящиеся в индексе и переносит их актуальный вид в репозиторий. Флаг `-m` указывает на сообщение коммита. Данная команда сохраняет изменения только локально. Для сохранения на GitHub необходимо использовать в дальнейшем другую команду.   

После этого заходим в наш профиль GitHub и создаем там новый репозиторий. Даем ему имя такое же, как и называется наша корневая папка.     

Далее необходимо связать локальный и удаленный репозиторий. После того, как мы создадим новый репозиторий в GitHub, перед нами откроется страница "Quick setup — if you’ve done this kind of thing before". Под этой записью есть возможность выбора между двумя протоколами: HTTPS и SSH. Разница между ними заключается в том, что при HTTPS нужно вводить токен при каждом push (или настраивать credential manager), а при SSH пароль не нужен вообще, ключ подхватывается автоматически. Стандартом является SSH, так как он является еще и более безопасным, по сравнению с HTTPS. Поэтому выбираем протоколо SSH (его заранее нужно настраивать) и копируем со страницы строку `git@github.com:NAME/NAME_REPOSITORY.git`. Скопированный текст должен применяться в команде `git remote add origin`. Вместе получится такая команда: `git remote add origin git@github.com:p1xelmind/tabular-bank-churn-logistic-regression.git`. Данная команда прописывает адрес удалённого репозитория в локальном Git. `origin` - это стандартное имя-алиас для основного удалённого репозитория.   

После этого прописываем команду `git branch`. Данная команда выводит список всех локальных веток репозитория. Текущая активная ветка помечается звёздочкой `*` и выделяется зелёным цветом. Если эта ветка называется `* master`, то переименовываем ветку, так как сейчас актуальное название `* main`. Переименование в main делается командой `git branch -M main`. Если команда `git branch` выдала `* main`, то можно без переименования сразу грузить на GitHub. Выгрузка на GitHub делается командой `git push -u origin main`.    

После этого в профиле GitHub появится новый репоизиторий, будет сформирован из текущей папки проекта. Чтобы обновлять проект, требуется использовать перечень команд: `git add file_name`, `git commit -m "commit"` и `git push`. В первой команде вместо `file_name` используем полное название файла, который мы модернизировали. Можно также использовать точку, чтобы перенести в индекс все файлы, но тогда будет проблема с написанием комментария к коммиту.   

Как **итог**, выпишим все команды и действия в правильной последовательности:   
`git init`    
`git add .`   
`git status`     
`git commit -m "commit"`   

Заходим на GitHub и создаем новый репозиторий.   

`git remote add origin git@github.com:NAME/NAME_REPOSITORY.git`    
`git branch`   

если `* master`, то:    
`git branch -M main`    
`git push -u origin main`    

если `* main`, то:   
`git push -u origin main`

#### Баланс классов (Class Balance)
Баланс классов - это соотношение количества объектов разных классов в целевой переменной. В данном случае целевой переменной является колонка `Exited`, которая имеет 2 класса: 0 (клиент остался) и 1 (клиент ушел). Логистическая регрессия (и большинство ML-моделей) по умолчанию предполагает, что классы представлены примерно поровну. Если в тренировочном датасете будет дисбаланс, то мы столкнемся с рядом проблем:  
- Модель "ленится". Если в датасете 80% остались, а 20% ушли, то модель может научиться просто предсказывать "0" для всех. Она получит точность (Accuracy) 80%, но не выявит ни одного ушедшего клиента. Такая модель будет бесполезна. 
- Смещение коэффициентов. Модель будет оптимизировать функцию потерь в пользу мажоритарного класса, так как ошибок на нем больше. Коэффициенты логистической регрессии будут смещены в сторону предсказания класса 0.  
- Неверные метрики. Accuracy становится обманчивой. Придется использовать ROC-AUC, Precision, Recall и F1-score, но даже они могут страдать, если модель никогда не видит достаточно примеров миноритарного класса во время обучения.

Как бороться с дисбалансом классов: 
- Использовать параметр `class_weight='balanced'` в LogisticRegression. Модель автоматически увеличит штраф за ошибки на миноритарном классе, компенсируя дисбаланс весами.
- Ресэмплинг. Либо удалить часть мажоритарного класса, но при этом потерять информацию. Либо синтезировать новые примеры миноритарного класса, но при этом увеличить риск переобучения. 
- Изменение порога классификации. По умолчанию порог 0.5. Можно снизить его до 0.3 или 0.4, чтобы модель чаще предсказывала "ушел".

Наиболее рекомендуемым способом для старта является первый вариант.   

Чтобы узнать количество значений "0" и "1" используем метод `value_counts()`, а чтобы узначть количество в процентном соотношении, воспользуемся параметром `normalize`, который при True преобразует значения в относительные частоты. По умолчанию он равен False:   
```
print(train['Exited'].value_counts())
print(train['Exited'].value_counts(normalize=True) * 100)
```
Можем также результат умножить на 100, чтобы посмотреть именно на процентное соотношение.

Проанлизируем получившиеся результаты. Шкала дисбаланса выглядит так: 
- `50/50 - 60/40` - нет дисбаланса или минимальный дисбаланс. Модель видит достаточно примеров обоих классов. Веса не нужны, они могут даже добавить лишний шум. Поэтому ничего исправлять не нужно.  
- `60/40 - 70/30` - слабый уровень дисбаланса. В этом случае можно воспользоваться балансировкой (`class_weight='balanced'`). Но можно начать и без нее, а если метрика (ROC-AUC/F1) проседает, то можно добавить, но часто разница минимальна.  
- `70/30 - 85/15` - умеренный уровень дисбаланса. В этом случае уже обязательно нужно добавлять балансировку. Модель уже заметно смещается к мажоритарному классу.  
- `85/15 - 95/5` - сильный уровнь дисбаланса. Нужно обязательно добавлять балансировку + анализ ошибок. Весов может быть недостаточно. Нужно смотреть Precision-Recall кривую (не только ROC-AUC), возможно настраивать порог классификации. SMOTE/undersampling - как эксперимент, но с осторожностью. 
- `>95/5` - крайне высокий уровень дисбаланса. `class_weight` часто не справляется. Нужны: oversampling (SMOTE/ADASYN), anomaly detection подход, специфичные функции потерь (Focal Loss), или сбор дополнительных данных миноритарного класса.

**Балансировка весов** (`class_weight='balanced'`) - это прием при котором мы меняем цену ошибки, чтобы скомпенсировать дисбаланс классов. Другими словами, балансировка делает ошибку на редком классе дороже. Представим, что у нас 100 клиентов. 80 из них остались клиентами банка, а 20 ушли. Модель при обучении хочет получить максимальный средний балл. Самая простая стратегия - считать, что все клиенты остались. В этом случае модель ошибется на всех тех, кто ушел, но в конечном итоге результат будет 80% точности. Когда мы применяем балансировку весов, то мы говорим модели, что ошибка на миноритарном классе (те 20 ушедших) стоит в 4 раза дороже, чем ошибки на мажоритарном классе. И модель будет вынуждена учиться различать разные классы. Но почему ошибка становится именно в 4 раза больше, откуда берется этот коэффициент? Он высчитывается делением мажоритарного класса на миноритарный: 79/21 ≈ 3.76. Так и получился этот коэффициент.    

**Ресэмплинг** - это физическое изменение обучающийся выборки до передачи ее в модель. *Undersampling* - удаление случайных объектов мажоритарного класса, пока соотношение не станет 50/50. Но 50/50 не является обязательным требованием, а лишь одна из возможных стратегий. В большенстве случаев нужно делать соотношение 80/20 + `class_weight='balanced'`. *Oversampling* - это синтез новых объектов миноритарного класса, пока соотношение не выровняется. Существует несколько методов oversamling'а. Рассмотри их:   
1) Random Oversampling - дублирует случайные существующие объекты миноритарного класса, но из-за того, что он создает точные копии, высок шанс переобучения.   
2) SMOTE - синтезирует новые объекты путем интерполяции между k ближайшими соседями миноритарного класса. Генерирует правдоподобные, но искусственные точки; может создавать шум в областях перекрытия классов. 
3) ADASYN - улучшенная версия SMOTE: генерирует больше синтетических объектов там, где модель уже ошибается. Адаптивнее, но сложнее и медленнее.
4) Borderline-SMOTE - генерирует объекты только на границе классов. Фокусируется на самой сложной зоне разделения. 

**Нюансы, которые влияют на выбор стратегии:**     
- Помимо процентов важно еще анализировать абсолютное количество данных. Если при 80/20 при 100k строк, 20% - это 20k строк миноритарного класса и в этом случае `class_weight='balanced'` сработает отлично, то в 80/20 при 1000 объектов, 20% - это всего 200 строк. В этом случае балансировки не хватит. Если минориторный класс >5000 объектов, то можно начинать с `class_weight='balanced'`, если же <1000 объектов, то сразу нужно думать о ресэмплинге.    
- Метрика определяет порог "достаточности". ROC-AUC оценивает ранжирование, а не абсолютные предсказания. Она более устойчива к дисбалансу, поэтому class_weight часто хватает до 90/10. F1/Precision/Recall зависят от порога 0.5. При дисбалансе >80/15 оптимальный порог почти никогда не равен 0.5. Здесь нужна не только балансировка, но и подбор порога через Precision-Recall curve.
- Бизнес-контекст. Если ложное срабатывание (False Positive) дорого (например, блокировка легальной транзакции), то лучше недообучить миноритарный класс, чем переобучить его через SMOTE. Если пропуск позитивного класса (False Negative) критичен (медицинская диагностика, мошенничество), то агрессивная балансировка и снижение порога оправданы.

**Разница между метриками:**    
В логистической регрессии метрики измеряют разные аспекты качества, и выбор между ними зависит не от вкуса, а от цены ошибки в реальной жизни. Метрик существует огромное множество, но ROC-AUC, F1, Precision, Recall покрывают 95% случаев. Рассмотрим каждую из них поподробнее:
- `ROC-AUC` отражает способность модели ранжировать клиентов по вероятности ухода, независимо от того, какой порог будет выбран. Это главная метрика для сравнения моделей. Если у Модели А ROC-AUC = 0.85, а у Модели Б = 0.79, то Модель А фундаментально лучше разделяет классы. То есть данная метрика отражает то, насколько модель хорошо различает классы. Если ROC-AUC = 1.0, то модель идеальна, она находит все объекты того класса, который нужен, и игнорирует все объекты класса, который не нужен. Если ROC-AUC = 0.5, то модель не лучше, чем случайное угадывание. Это главная метрика оценки. Если она низкая, никакой настройкой модель не улучшишь.   
- `Precision` отражает долю правильных предсказаний. Она нужна тогда, когда ресурсы ограничены и ложные срабатывания стоят денег. Данная метрика говорит, насколько вообще можно доверять модели. Если из 10 срабатываний на нужный класс, правльных было только 8, то Precision = 80%. Чем выше точность, тем меньше ложных срабатываний. Данная метрика нужна тогода, когда каждая ошибка стоит дорого и лучше пропустить объект нужного класса, но при этом допустить меньше ошибок.  
- `Recall` отражает полноту срабатывания модели. Она помогает понять, сколько мы вообще "увидели" объектов нужного класса из всех существующих объектов. Если объектов класса 1 всего 100 в датасете, а модель "увидела" только 80 из них, то Recall = 80%. Данная метрика крайне важна, если мы анализируем какие-то жизненно важные ситуации, при которых лучше ошибиться много раз, но найти все объекты требуемого класса.   
- `F1-Score` отражает сбалансированный показатель `Precision` и `Recall`. При высоком Recall и низком  Precision или наоборот, при высоком Precision и низком Recall -> F1-Score будет низким. Но данный показатель будет высоким, когда Recall и Precision имеют приемлемый уровень. То есть данная метрика отображает компромис между "точностью" и "полнотой".  

**True/False Positive/Negative:**    
- `TP (True Positive)` - Истинно положительный. Модель сказала "1", и это реально "1". Это корректная работа модели.  
- `TN (True Negative)` - Истинно отрицательный. Модель сказала "0", и это реально "0". Это корректная работа модели.  
- `FP (False Positive)` - Ложно положительный. Модель сказала "1", но в реальности "0". Это некорректная работа модели.  
- `FN (False Negative)` - Ложно отрицательный. Модель сказала "0", но в реальности "1". Это некорректная работа модели.  

**Порог:**    
Логистическая регрессия предсказывает не класс (0 или 1), а вероятность принадлежности к классу "1". Результат работы модели - число от 0 до 1. Если модель выдает результат 0.8, то она "уверена" на 80%, если результат модели 0.2, то модель "уверена" на 20%. Чтобы превратить эту вероятность в конкретный ответ (в контексте данного задания - уйдет клиент из банка или нет), нам потребуется порог. Стандартный порог равен 0.5. Если $P \ge 0.5$, то опрделеяем это как класс "1" (клиент ушел), а если $P < 0.5$, то определяем этот результат как "0" (клиент остался). Порог 0.5 - это просто договоренность по умолчанию, его можно менять при необходимости. *Метрики Precision/Recall зависят от порога, а ROC-AUC — нет (она оценивает модель при всех возможных порогах сразу)*.  

Как видно из шкалы, мы имеем дело с умеренным уровнем дисбаланса и при реализации модели логистической регрессии нужно будет реализовать балансировку весов с помощью `class_weight='balanced'`.   

#### Распределение целевой переменной по признакам
Вторым пунктом в чек-листе является "распределение целевой переменной по признакам". Этот пункт должен дать нам понимание, есть ли вообще связь между признаками и целевой переменной.В этом пункте необходимо выяснить 4 нюанса: 
1) Проверка пригодности признака. Если распределение `Age` одинаковое для ушедших и оставшихся, значит возраст не несёт сигнала. Такой признак - шум, его можно удалить или не ожидать от него пользы. Пригодным признаком называют тот, распределение которого статистически значимо различается между классами Exited=0 и Exited=1.  
2) Понимание характера связи. Логистическая регрессия предполагает линейную зависимость между признаком и логарифмом шансов. Если ты видишь, что уходят и молодые (<25), и пожилые (>60), а средние остаются - связь нелинейная. Модель это не выучит без трансформации (например, биннинга возраста).
3) Идеи для Feature Engineering. Если, к примеру, оттток клиентов из Германии выше, чем во Франции, значит, возможно, стоит создать новый признак. Если видно, что баланс у ушедших ниже, значит, возможно, важен не сам баланс, а отношение баланса к зарплате.   
4) Интерпретируемость. Когда модель обучится, ты сможешь объяснить бизнесу: "Клиенты старше 50 уходят чаще, потому что..." - и это будет подтверждено данными, а не только коэффициентами модели.  

**Проверка на пригодность признака:**    
Для того, чтобы определить пригодность признака необходимо провести несколько манипуляций. Сначала определить пригодность признаков визуально (с помощью построения графиков), а потом и статистически. Методы построения графика и статистического анализа будут отличаться в зависимости от того, к какому типу относится признак: числовому или категориальному (к категориальным признакам относят как номинальные, так и порядковые).

На графике видим 2 синих прямоугольника. Каждый из этих прямоугольников показывает, где находится интерквартильный размах (50% всех клиентов). Boxplot (ящик с усами) строится независимо для каждой группы на оси X. То есть статистика считается отдельно для класса 0 и отдельно для класса 1. Они не смешиваются. В данном случае у нас 2 отдельные очереди клиентов: "0" - те, что остались и "1" - те, что ушли. Для каждой очереди люди выстроены по росту кредитного скоринга (от самого низкого до самого высокого). Область под коробкой - это 25-й перцентиль. Это значение скора, ниже которого находятся самые "низкие" 25% клиентов этой конкретной группы. Выше коробки находится 75% перцентиль, то есть там находятся клиенты, которые имеют рейтинг выше 75% от всего диапозона рейтинга. А внутри этих прямоугольникох находится ровно 50% всех клиентов.    

То есть если в классе "0" 1000 клиентов, то в прямоугольнике будет находиться ровно 500 клиентов, под прямоугольником будет находиться область с 250 клиентами с самым низким кредитным рейтингом, а над прямоугольником будет находиться обалсть с 250 клиентами уже с самым высоким рейтингом. Области над и под прямоугольником ограничены короткими горизонтальным полосками. Иногда за пределами этих полосок могут быть точки (круги) - это выбросы, клиенты, которые не попали в, так называемые, усы. Обычно их 1-2% от общего числа. В нашем случае, есть выбросы ниже усов. Это небольшое количество людей, которые имеют крайне низкий рейтинг. Горизонтальная линия внутри этих коробок - это медиана или же по-другому 50% перцентиль. Она делит выборку пополам по количеству людей, но НЕ делит прямоугольник пополам по высоте. Потому что данные распределены неравномерно.   

Рассмотрим класс "0" на нашем графике. Коробка занимает диапазон ~600 - 710 единиц рейтинга. Высота ~110 единиц. Усы тянутся от ~430 до ~850 (общий размах 420 единиц). Для удобства, примем 1000 за количество клиентов в классе "0". В этом случае, в коробке 500 человек, под коробкой 250 человек, над коробкой тоже 250 человек. Для распределения этих всех клиентов выставляют в ряд по возрастанию. Под Q1 (25%) - это 250 человек с самым низким рейтингом, и судя по графику, это 250 человек с рейтингом от 430 до 600. Над Q3 (75%) - это 250 человек с самым высоким рейтингом. И судя по графику, это люди с рейтингом от 710 до 850. Остальные 500 имеют рейтинг от 600 до 710, что говорит о том, что они расположены близко друг к другу на координатной прямой. Горизонтальная линия в прямоугольнике - это 500-й человек из 1000 или же 250-й из 500 (из коробки). И он имеет примерно 670 рейтинга. И то, что горизонтальная линия расположена не четко по центру квадрата - это абсолютно нормально, потому что данные распределены неравномерно. Если медиана ниже центра коробки, значит большая часть людей "кучкуется" внизу (плотность в нижней части больше, чем в верхней). Если же медиана выше центра коробки, как в нашем случае, значит большая часть людей сосредоточена вверху (так как на 250 людей выделено меньше площади, соответственно, плотность выше).   

Теперь сделаем вывод по графику. Мы видим, что:   
1) Горизонтальная линия (медиана) внутри коробки для `Exited=1` и `Exited=0` находится практически на одном уровне (~655-660). Это значит, что что "типичный" ушедший клиент и "типичный" оставшийся клиент имеют одинаковый кредитный рейтинг.  
2) Идентичный размах. Сами синие прямоугольники имеют одинаковую высоту и расположены на одной высоте. То есть "средние 50%" клиентов в обеих группах имеют одинаковый диапазон рейтингов (~600–710).
3) Идентичные усы и выбросы. Вертикальные линии (усы) и точки-выбросы внизу также совпадают по положению и количеству. Распределение крайних значений тоже одинаковое.

Соответственно, на графике нет никаких существенных различий между классами. А знаит он не влияет на итоговый результат: уйдет клиент из банка или нет. И в этом случае у нас есть 2 пути: либо удалить признак, либо оставить и использовать регууляризацию L1/L2, которая сама обнулит вес признака. Лучше всего оставить этот признак на данный момент, так как визуальная оценка - субъективна. То, что на boxplot не видно различий, не гарантирует, что модель тоже их не увидит. Иногда слабые сигналы проявляются только во взаимодействии с другими признаками (например, CreditScore + Age). Если же модель посчитает, что признак все же бесполезен, она сама обнулит или минимизирует ему вес с помощью регуляризации. И на этапе EDA признаки, как правило, не удаляются (только если они не являются явным мусором, как id или CustomerId).    

**Статистический анализ признака CreditScore:**   
Теперь, когда мы визуально провели анализ признака и увидели, что он не влияет на целевую переменную, можно переходить к статистическому анализу, чтобы убедиться в бесполезности признака.    
Для числовых признаков используется `Mann-Whitney U test`. Это непараметрический статистический тест, который отвечает на один вопрос: "Являются ли два независимых распределения статистически различными?".   

**Статистически различные распределения - это разница между множествами, которая носит систематический характер, а не случайный.** Представим 2 коробки с шарами: "A" и "B". В коробке A шары, которые пронумерованы от 1 до 100, а в коробке B шары, которые пронумерованы от 90 до 200. Если мы вытащим по одному шару из каждой коробки, то числа будут разными - это просто случайность. А если мы вытащим по 1000 шаров из каждой коробки, то увидим системную разницу: В коробке "B" числа как правило будут больше, чем числа в коробке "A". Это и есть "различные распределения". А теперь представим, что в обоих коробках шары пронумерованы от 1 до 100, но в коробке "А" случайно попались два шара с номером 95, а в коробке "B" - один шар с номером 96. Разница есть? Да. Но она случайная. Если перемешать и вытянуть заново, эта разница исчезнет или поменяется знак. **Статистическая значимость** - это ответ на вопрос: "Эта разница реальна и системна, или это просто случайный шум выборки?"    

Почему мы используем именно `Mann-Whitney U test`, а не корреляцию или t-test? Дело в том, что корреляция и t-test требует, чтобы данные были распределены **нормально** (как колокол). Но в реальном мире данные редко бывают идеально нормальными. Они могут быть скошены, иметь выбросы или странную форму. `Mann-Whitney U test` не требует нормальности. Он работает с рангами, а не с сырыми значениями.      

Но что такое **нормальное распределение?** Это знаменитая "кривая Гаусса" или "колокол". Представим рост всех мужчин в стране. Большенство мужчин имеет средний рост - 175 см, это пик "колокола". Очень высоких и очень низких мало, это пологие хвосты слева и справа. Распределение симметрично относительно среднего. И многие статистические тесты (например t-test) математически выведены предполагая, что данные выглядят как этот колокол. Если данные скошены (например, зарплаты: большинство получают мало, но есть несколько миллионеров, растягивающих правый хвост), t-test даст неверный результат. `Mann-Whitney U test` не требует колокола. Ему всё равно, как выглядят данные. Он работает даже если распределение похоже на букву U, L или вообще хаос. Именно поэтому он безопаснее для реальных данных. Но если же мы решили использовать t-test, то нужно было бы убедиться, что данные распределены нормально у каждого CreditScore (и для класса 0, и для класса 1). Если хотя бы одна колонка (из двух) CreditScore скошена, то t-test не применим.     

Выбор стратегии зависит не только от распределения данных у исследуемого признака, но и от типа целевой переменной. Всего есть 3 типа целевых переменных в табличных данных: 
1) Бинарная целевая переменная (для задач бинарной классификации) - предсказываем один из ДВУХ классов. 
2) Категориальная целевая переменная (для задач мультаклассовой классификации) - предсказываем один из N классов.
3) Числовая целевая переменная (для задач регрессии) - предсказываем конкретное число.

Таким образом, если у нас: 
- Бинарная целевая переменная и данные нормальные, то используем `t-test`
- Бинарная целевая переменная и данные не нормальные или не знаем какие они, то используем `Mann-Whitney U test`
- Числовая целевая переменная и данные нормальные, то используется `корреляция Пирсона` 
- Числовая целевая переменная и данные не нормальные или не знаем какие они, то используем `корреляцию Спирмена`
- Категоральная целевая переменная (признак числовой) и данные нормальные, то используем `ANOVA`
- Категоральная целевая переменная (признак числовой) и данные не нормальные или не знаем какие они, то используем `Kruskal-Wallis`
- Категоральная целевая переменная (признак категориальный), то всегда используем `Chi-square`

Теперь, когда мы полностью опеределилсь с методом анализа - `Mann-Whitney U test`, приступим к его реализации. Для этого будем использовать функцию `mannwhitneyu()` из `scipy.stats`. Импортируем функцию в коде: 
```
from scipy.stats import mannwhitneyu

credit_score_stayed = train[train['Exited'] == 0]['CreditScore']
credit_score_exited = train[train['Exited'] == 1]['CreditScore']
```
Далее создаем 2 переменных, которые являются Series со значениями колонки CreditScore. Разберем, как это получается. Мы для этого используем **булеву маску**. Булева маска - это объект типа Series, который заполнен значениями True/False, которые являются следствием сравнения изначального Series с каким-то значением. И тут булева маска выглядит так: `train['Exited'] == 0`. Из DataFrame берется Series 'Exited' и каждое значение этой колонки сравнивается с нулем. Как итог, мы получим Series, заполненный значениями True/False.    
Далее применяем эту булеву маску к DataFrame - train. Индексы маски сопоставляются с индексами всего DataFrame и как итог, мы получаем DataFrame, в котром остались только те строки, которые имели True в маске. И в такой таблице в колонке 'Exited' остались только значения '0', при этом структура изначального DataFrame сохраняется, в нем остались все те же колонки, что и были в нем изначально.     
После этого применяем к DataFrame `['CreditScore']`, чтобы получить Series, состоящий из значений 'CreditScore'. Как итог, в переменной `credit_score_stayed` будет храниться перечень значений кредитного рейтинга у клиентов, которые остались.     
Такой же механизм и у переменной `credit_score_exited`. Только в ней будет храниться Series со значениями кредитного рейтинга клиентов, которые ушли из банка.    

После этого используем эти Series в функции `mannwhitneyu()`:   
```
stat, p_value = mannwhitneyu(
    credit_score_stayed, 
    credit_score_exited, 
    alternative='two-sided'
)
```
Данная функция возвращает 2 переменные: stat и p_value. Принимает на вход 2 объекта Series, которые мы инициализировали ранее и параметр `alternative`. Разберем то, как работает данная функция.    

Итак, функция не берет колонку признака и целевую переменную, она берет 2 массива значений одного и того же признака, уже разделенные по классам. Массив 'A' - CreditScore клиентов, которые остались (класс 0); массив 'B' - CreditScore клиентов, которые ушли (класс 1).    
Далее реализуется сортировка. Оба массива объединяется в один и все значения сортируются от меньшего к большему.    
После этого всем значениям присваиваются ранги. Ранг - это просто позиция элемента в массиве, только начинается присваивание с 1 (а не с 0). Если вдруг 2 элемента в этом массиве имеют одинаковое значение, то им присваивается средний ранг. То есть вместо присваивания ранга 3 и 4, обоим присваивается ранг 3.5. Это назвается "поправка на связи".    
После этого вычисляется U-статистика на основе сумм рангов. U-статистика отвечает на вопрос: "Сколько раз значение из класса 0 меньше, чем значение из класса 1?". То есть мы берем каждое значение из класса "0" (массив A) и сравниваем его с каждым значением из класса "1" (массив B). Для каждой пары (значение из A, значение из B) мы проверяем: если A < B, добавляем +1. Если A > B, добавляем 0. Если равны — +0.5.    

Для удобства восприятия, представим, что у нас всего 6 клиентов:    
Класс 0 (остались): [600, 650, 700]     
Класс 1 (ушли): [620, 680, 750]     

Объединяем их в один массив и сортируем:     
[600, 620, 650, 680, 700, 750]      

Причисляем каждому значению ранг:   
| Значение | Ранг | Из какого класса? |
| :--- | :---: | :--- |
| 600 | 1 | Класс 0 |
| 620 | 2 | Класс 1 |
| 650 | 3 | Класс 0 |
| 680 | 4 | Класс 1 |
| 700 | 5 | Класс 0 |
| 750 | 6 | Класс 1 |

И начинаем сравнивать:    
600 < 620 ✅ → +1     
600 < 680 ✅ → +1     
600 < 750 ✅ → +1     
650 < 680 ✅ → +1     
650 < 750 ✅ → +1     
700 < 750 ✅ → +1    
650 > 620 ❌ → 0    
700 > 620 ❌ → 0    
700 > 680 ❌ → 0    

Как итог `U = 6` (из 9 возможных). Максимально возможное `U = n₁ × n₂ = 3 × 3 = 9`. Если бы все значения класса 0 были меньше всех значений класса 1, U было бы равно 9 (идеальное разделение). Если бы распределения были идентичны, U было бы около 4.5 (половина от 9).    

Значение U - это просто сырой счетчик парных значений. Само по себе оно ни о чем не говорит. Значение U имеет смысл только в контексте размера выборок и p-value.    

p-value - это вероятность того, что данные получились случайно, при условии, что на самом деле никакой разницы между группами нет.    
Представим монету. Подбросим ее 10 раз и получим 9 орлов. Нулевая гипотеза: "монетка честная, орел и решка равновероятны". P-value отвечает на вопрос: "Какая вероятность получить 9 орлов из 10 бросков, если монетка действительно честная?". Эта вероятность ~0.01 (1%), как следствие, сомневаемся, что монетка действительно честная.   

В контексте задания: **Нулевая гипотеза: "Распределение CresitScore у ушедших и оставшихся одинаковы"**. Это нулевая гипотеза по умолчанию, мы всегда начинаем именно с нее. То есть чтобы мы не исследывали, изначально мы предполагаем, что распределения одинаковы. P-value отвечает: "Какова вероятность получить такую U-статистику (или более экстремальную), если распределения действительно одинаковы (если мы придерживаемся нашей нулевой гипотезы)?". Если p-value будет 0.0000, значит, что такая U практически нереальна при одинаковых распределениях (при нулевой гипотезе), а следовательно, различие реально. Если же p-value = 0.45 (45% вероятность), то такая U легко получается при нулевой гипотезе, следовательно, различий нет и распределение одинаково. В этом случае мы считаем, что у нас недостаточно отвергать нулевую гипотезу (то есть мы не подтвержжаем ее, а лишь не можем опровергнуть).   

Итак, возвращаемся к коду. Как мы помним, данная функция возвращает 2 значения - `stat` и `p_value`. Так вот `stat` - это значение U, а `p_value`, соответственно, p-value.    
Помимо двух Series со значениями CreditScore для разных классов, мы еще передаем аргумент `alternative`. Данный параметр определяет направление гипотезы, что именно мы проверяем. Он отвечает на вопрос: "в какую сторону мы ищем различия?". Имеет 3 возможных значения. `'two-sided'` - нулевая гипотеза "распределения идентичны". Это значение - дефолт для EDA. `greater` - группа 1 ≤ группы 2. Используется только если есть априорное ожидание направления. `less` - группа 1 ≥ группы 2. Также только в том случае, если есть априорные ожидания.    

Как уже было выше сказано, для EDA используется значение `'two-sided'`, так как наша задача заключается в том, чтобы узнать, есть ли вообще какие-то различия в распределении.   

И после этого выводим на полученные значения:   
```
print(f"U-stat: {stat:.2f}")
print(f"P-value: {p_value:.4f}")
```
В контексте данного задания получатся вот такие значения на выходе:    
```
U-stat: 2365349854.50
P-value: 0.0000
```
Итак, мы видим, что p-value равен 0.0000 при аргументе `alternative='two-sided'`. То есть вероятность получить такую U-статистику при условии, что распределения одинаковы - околонулевая.     

Но не будем спешить. Вспомним, что у нас в датасете 165.000 строк. При таком объёме данных тест улавливает даже микроскопическую разницу в медианах (например, 3–5 баллов), которая на boxplot выглядит как "коробки почти совпадают". P-value отвечает на вопрос "Есть ли различие?", но НЕ отвечает на вопрос "Насколько оно велико?". При больших выборках ответ на первый вопрос почти всегда "да", даже когда второй ответ - "ничтожно мало".   

Как в этом случае интерпетировать результат статистической проверки? При p-value равном 0, но идентичных "коробках" на графиках и незначительных отличиях в медианах - **различие статистически значимо, но практически бесполезно для модели**.
